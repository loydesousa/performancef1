```python
import pandas as pd

df = pd.read_csv("Formula1_2025Season_RaceResults.csv")

df["Position"] = pd.to_numeric(df["Position"], errors="coerce")

# Calcular estatísticas por piloto
pilotos_stats = df.groupby("Driver")["Position"].agg(
    total_corridas="size",
    # Para cada piloto, conta quantas linhas/participações ele tem (quantas corridas aparecem no CSV).

    vezes_top5=lambda x: (x <= 5).sum(),
    # Comparações com NaN resultam em False, então corridas sem posição válida não contam como top 5.
)

# Criar coluna de vezes fora do top 5
pilotos_stats["vezes_fora_top5"] = (
    pilotos_stats["total_corridas"] - pilotos_stats["vezes_top5"]
)

# Ordenar pilotos mais consistentes
ranking_consistencia = pilotos_stats.sort_values(
    by=["vezes_fora_top5", "vezes_top5"],
    # Critério 1: menos vezes fora do top 5 (melhor é menor).
    # Critério 2: (desempate): mais vezes no top 5 (melhor é maior).
    ascending=[True, False]
    # Define a ordem para cada critério: crescente para "fora_top5" e decrescente para "vezes_top5".
)
#Ranking Top 5
top5_consistentes = ranking_consistencia.head(5)
print("Top 5 pilotos mais consistentes no top 5:")
print(top5_consistentes)
```
